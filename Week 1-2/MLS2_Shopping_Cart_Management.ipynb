# **Simplified Shopping Cart Management System for a Fruit Seller**

A local fruit seller faces significant challenges in managing the billing process for customers. Currently, the seller uses a manual method, which is time-consuming, prone to errors, and lacks efficiency. This can lead to customer dissatisfaction and potential revenue loss due to mismanagement or billing inaccuracies.

To address this, the fruit seller is exploring the implementation of a basic Shopping Cart Management System as an initial test to improve the shopping experience. The system does not need to be overly complex but should provide essential functionalities to streamline operations and improve customer experience.


<hr>

**Business Requirements:**

The proposed Shopping Cart Management System should include the following key features:

1. **Add Products to the Shopping Cart:** Seller should be able to add items to the shopping cart. The system will pull product details such as name, price directly from the inventory. This ensures that the cart remains synchronized with the available products.

2. **Remove Unwanted Items from the Shopping Cart:** Seller should be able to remove items from the shopping cart if they are no longer needed. This feature provides flexibility and reduces errors in the billing process.

3. **Calculate the Total Bill:** The system should automatically calculate the total cost of the items in the cart. This will eliminate manual calculations and ensure accurate billing.

## 2. Write Functions to add Items to Cart

### Add items to cart

```python
# Initailizing Your_Cart here
Your_Cart = []
```

```python
"""
Write a Python function add_item_to_cart(item_id, Inventory) that adds an item from the Inventory (using item_id - 1 for zero-based indexing) to a shopping cart "Your_Cart".
Display a confirmation message with > and < decorations confirming the item has been added and the cart updated. If the item_id is invalid, display an appropriate error message.
"""

def add_item_to_cart(item_id, Inventory):
    '''
    Add the Items_id to your cart and display the message.
    '''
    try:
        item = Inventory[item_id - 1]
        Your_Cart.append(item)
        print(f"{'>'*10} The item '{item[1]}' has been added to the cart, and your shopping cart has been updated. {'<'*10} \n\n")
    except IndexError:
        print(f"Invalid item ID: {item_id}")
```

```python
## Add Test Case

# Test Cases
# Test 1: Add a valid item
print("Test 1: Adding a valid item")
add_item_to_cart(1, Inventory)
print(f"Expected: 1 item in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

# Test 2: Add another valid item
print("Test 2: Adding another valid item")
add_item_to_cart(2, Inventory)
print(f"Expected: 2 items in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

# Test 3: Try to add an invalid item
print("Test 3: Adding an invalid item")
add_item_to_cart(99, Inventory)
print(f"Expected: Error message and 2 items in cart, Found: {len(Your_Cart)} item(s) in cart.\n")
```

### Remove items from cart

```python
"""
Write a Python function remove_item(cart_index, cart_items) that removes an item from the cart_items list at the given cart_index.
Display a confirmation message with > and < decorations if the item is removed, or an error message if the cart_index is invalid or the cart is empty.
"""

def remove_item(cart_index, cart_items):
    '''
    Remove item with given item_id and display the message.
    '''
    try:
        removed_item = cart_items.pop(cart_index - 1)
        print(f"{'>'*10} The item '{removed_item[1]}' has been removed from the cart, and your shopping cart has been updated. {'<'*10} \n\n")
    except IndexError:
        print("Invalid cart index or cart is empty.")
```

### Total cost of items in cart

```python
"""
Write a Python function `total_cost(cart_items)` that calculates the total cost of all items in the `cart_items` list (each item being a tuple where the price is the fourth element) and prints the total cost in a formatted message.
Display the total value using ~ symbols as decorative separators above and below the total. Center-align the text displaying the total cost within the decorations.
"""

def total_cost(cart_items):
    '''
    Calculate the total cost of the items in the shopping cart and display the total value.
    '''
    total = 0
    for item in cart_items:
        total += item[2]
        # item_price = item[2]
        # total = total + item_price

    print("~"*40)
    print(f"Total cost : {total}".center(40) )
    print("~"*40, "\n\n")
```

### Display the complete cart

```python
"""
Write a Python function display_cart(Your_Cart) that checks if the Your_Cart list is not empty. If the cart contains items, it prints a formatted list of the items (with their index, name, quantity and price per kg).
The function should display a decorative header and footer, with the total cost of all items displayed at the end. If the cart is empty, it prints a message saying the cart is empty.
"""

def display_cart(Your_Cart):
    # Displaying your cart items in decorative format if the cart is not empty
    if Your_Cart:
        print("+" * 50)
        print(f"{'Your Cart'.center(40)}")
        print("+" * 50)
        print(f"{'Index':<10} {'Name':<20} {'Quantity':<10} {'Price Per KG':<10}")
        print("-" * 50)

        unique_items = set(Your_Cart)  # Get unique items in the cart
        print(unique_items)
        total_cost = 0
        for index, item in enumerate(unique_items):
            quantity = Your_Cart.count(item)  # Calculate quantity of each item
            price_per_kg = item[2]
            total_price = price_per_kg * quantity
            print(f"{index+1:<10} {item[1]:<20} {quantity:<10} ${total_price:<10.2f}")
            total_cost += total_price
        print("-" * 50)
        print(f"{'Total Cost':<40} ${total_cost:<10.2f}")  # Display grand total

    else:
        print(f"{'>'*10} Your cart is empty !!!, Why not buy Fruits from us ... {'<'*10} \n\n")
```

## 3. Main Method

```python
""""
Write a Python program with a loop for a shopping cart system.
Display the cart using display_cart(Your_Cart) and prompt the user to:
  1. Add items (add_item_to_cart)
  2. Remove items (remove_item)
  3. Calculate total cost (total_cost)
  4. Exit the program (option 9)
The program should handle user input for these operations and continue prompting the user until they choose to exit.
For each action, prompt the user with appropriate messages and handle invalid input with error messages.
"""

def main():
  while True:
    display_cart(Your_Cart)

    try:
      user_choice = int(input(f"Choose an operation: \n{'-'*50} \n1. Add items to cart \n2. Calculate total cost  \n3. Remove an item \n---\n9.EXIT \n{'-'*50} \nEnter your choice: "))
    except ValueError:
      print("Invalid input. Please enter a number.")
      continue  # Go back to the beginning of the loop

    if(user_choice == 1): # If the choice is to add items to the cart
      item_id = int(input("Enter the Product ID of the item you wish to add to the cart : ")) # Prompt for the number of item IDs
      add_item_to_cart(item_id, Inventory) # Call the function to add items to the cart

    elif(user_choice == 2): # If the choice is to calculate the total cost
      total_cost(Your_Cart) # Call the function to calculate the total cost

    elif(user_choice == 3): # If the choice is to remove an item from the cart
      cart_index = int(input("Please enter the Product ID of the item you want to remove from the cart : ")) # Prompt for the Index of the item to be removed
      remove_item(cart_index,Your_Cart) # Call the function to remove the item

    elif(user_choice == 9): # Option to Exit
      break

    else:
        print("Invalid choice") # If the choice is invalid, print an error message


if __name__ == "__main__":
  main()
```

```python
"""
Create a Gradio interface for the complete code of the above case study.
The interface should allow users to add items to the cart, remove items, display the cart, and calculate the total cost interactively.
"""

from gradio import Blocks, Number, Button, Textbox

Inventory = [
    (1, "Apples", 250),
    (2, "Cherry", 650),
    (3, "Chickoo", 50),
    (4, "Grapes", 90),
    (5, "Mangoes", 150),
]

Your_Cart = []

def add_item_to_cart(item_id, Inventory):
    """
    Adds an item from the Inventory to the shopping cart.
    """
    try:
        item = Inventory[item_id - 1]
        Your_Cart.append(item)
        return f"{'>'*10} The item '{item[1]}' has been added to the cart. {'<'*10} \n\n"
    except IndexError:
        return f"Invalid item ID: {item_id}"

def remove_item(cart_index, cart_items):
    """
    Removes an item from the shopping cart.
    """
    try:
        removed_item = cart_items.pop(cart_index - 1)
        return f"{'>'*10} The item '{removed_item[1]}' has been removed from the cart, and your shopping cart has been updated. {'<'*10} \n\n"
    except IndexError:
        return "Invalid cart index or cart is empty."

def display_cart(Your_Cart):
    """
    Displays the items in the shopping cart.
    """
    if Your_Cart:
        cart_str = ""  # Initialize an empty string to store the cart display
        cart_str += "+" * 50 + "\n"
        cart_str += f"{'Your Cart'.center(40)}" + "\n"
        cart_str += "+" * 50 + "\n"
        cart_str += f"{'Index':<10} {'Name':<20} {'Quantity':<10} {'Price Per KG':<10}" + "\n"
        cart_str += "-" * 50 + "\n"
        unique_items = set(Your_Cart)  # Get unique items in the cart
        total_cost = 0
        for index, item in enumerate(unique_items):
            quantity = Your_Cart.count(item)
            price_per_kg = item[2]
            total_price = price_per_kg * quantity
            cart_str += f"{index+1:<10} {item[1]:<20} {quantity:<10} ${total_price:<10.2f}" + "\n"
            total_cost += total_price
        cart_str += "-" * 50 + "\n"
        cart_str += f"{'Total Cost':<40} ${total_cost:<10.2f}"  # Display grand total
        return cart_str  # Return the string representation of the cart
    else:
        return f"{'>'*10} Your cart is empty !!!, Why not buy Fruits from us ... {'<'*10} \n\n"

def total_cost(cart_items):
    """
    Calculates the total cost of all items in the shopping cart.
    """
    total = 0
    for item in cart_items:
        total += item[2]  # Assuming quantity is 1 for now
    return f"~"*40 + "\n" + f"Total cost: {total}".center(40) + "\n" + "~"*40 + "\n\n"

def gradio_interface(item_id, cart_index):
    """
    Handles adding/removing items, displaying the cart, and calculating the total cost.

    Args:
        item_id (int): The product ID to add to the cart (if any).
        cart_index (int): The product ID to remove from the cart (if any).

    Returns:
        tuple: A tuple containing the updated cart display, add message, remove message, and total cost.
    """

    # Add item
    add_msg = ""
    if item_id:
        add_msg = add_item_to_cart(item_id, Inventory)

    # Remove item
    remove_msg = ""
    if cart_index:
        remove_msg = remove_item(cart_index, Your_Cart)

    # Display the cart and calculate total cost
    cart_display = display_cart(Your_Cart)
    cost_msg = total_cost(Your_Cart)

    return cart_display, add_msg, remove_msg, cost_msg

# Gradio interface creation
def launch_gradio():
    """
    Launches the Gradio interface for the fruit shopping cart system.
    """

    with gr.Blocks() as demo:
        gr.Markdown("### Fruit Shopping Cart System")

        with gr.Row():
            item_id_input = gr.Number(label="Enter Product ID to Add to Cart", interactive=True)
            cart_index_input = gr.Number(label="Enter Index ID to Remove from Cart", interactive=True)

        with gr.Row():
            add_button = gr.Button("Add to Cart")
            remove_button = gr.Button("Remove from Cart")

        cart_display_output = gr.Textbox(label="Your Cart", interactive=False)
        add_message_output = gr.Textbox(label="Add Item Message", interactive=False)
        remove_message_output = gr.Textbox(label="Remove Item Message", interactive=False)
        total_cost_output = gr.Textbox(label="Total Cost", interactive=False)

        # Add item button click behavior
        add_button.click(
            gradio_interface,
            inputs=[item_id_input, gr.Number(visible=False, value=0)],  # Pass a hidden Number component for cart_index
            outputs=[cart_display_output, add_message_output, remove_message_output, total_cost_output]
        )

        # Remove item button click behavior
        remove_button.click(
            gradio_interface,
            inputs=[gr.Number(visible=False, value=0), cart_index_input],  # Pass a hidden Number component for item_id
            outputs=[cart_display_output, add_message_output, remove_message_output, total_cost_output]
        )
    demo.launch()

# Run the Gradio interface
if __name__ == "__main__":
    launch_gradio()
```
