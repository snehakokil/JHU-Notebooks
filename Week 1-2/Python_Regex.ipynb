**3. RegEx Module**

**Example - 1**

The below code checks if the string starts with "I" and ends with "coding". If yes, print "YES! We have a match!" and if no print "No match".

```python
import re
txt = "I love coding"
x = re.search("^I.*coding$", txt)
if x:
  print("YES! We have a match!")
else:
  print("No match")
```

**4. RegEx Functions**

The set of functions provided by the re module can be used to search for a string match.

Lets see various functions provided by the re module to interact with regex in Python.




**findall()**

findall() is a method in Python's re module that searches a string for all occurrences of a specified pattern and returns them as a list of strings. It can be used to locate every instance in a string where a regular expression matches.

**Example**

Return a list containing every occurrence of "on"

```python
import re
#Return a list containing every occurrence of "on":
a = "Python Programming is one of the best"
b = re.findall("on", a)
print(b)
```

**search()**

search() is a function in Python's regular expression module that searches for the first occurrence of a pattern in a string and returns a match object.

If no match is discovered, it returns None. The search() method can be used to locate a pattern anywhere in the string because it searches the entire string for the pattern.


**Example**

Searching white spaces 


```python
#search()
import re
a = "Python Programming is one of the best"
b = re.search("\s", a)
print("The first white-space character's position:", b.start()) 
```

**split()**

A string can be divided into a list of substrings using the split() method in the Python regular expression package.

When the pattern is met, the method splits the string into a list, which is what is returned.


**Example**

Split the words as and when white space is encountered 

```python
import re
a = "Python Programming is one of the best"
b = re.split("\s", a)
print(b)
```

**sub()**

The sub() method in the regular expression module of Python allows you to substitute a string for one or more matches of a pattern.

**Example**

Replace all white-space characters with the digit "5"

```python
#sub()
import re
#Replace all white-space characters with the digit "5":
a = "Python Programming is one of the best"
b = re.sub("\s", "5", a)
print(b)
```

**5. Metacharacters**

The re module's functions use metacharacters, which are fundamental building blocks of regular expressions, to define patterns that match particular strings in a text.

Here is a list of some of the most commonly used metacharacters in regular expressions:




**[ ]	- Square brackets** 

To define a character class in regular expressions, use square brackets ([]). The regular expression engine will try to match a character from a character class against a single character from the input string.

A single character that is either a "a," "b," or "c" will match the regular phrase [abc], for instance. So, the character "c" will match if you use this regular expression to look for a match in the string "cat" since "c" is one of the characters defined in the character class.





**Example**

Find all lower case characters alphabetically between "a" and "d".

```python
import re
a = "coding is the best"
#Find all lower case characters alphabetically between "a" and "d":
b = re.findall("[a-d]", a)
print(b)
```

**\  Backslash**

Regular expressions can treat metacharacters as literal characters by escaping them with the backslash (). 

When you need to match a specific character that also happens to be a metacharacter, like a dot (.) or an asterisk (*), we can use this technique. 

We can tell the regular expression engine that a metacharacter should be interpreted as a regular character and not as a special character with its usual meaning by adding a backslash before it.

**Example**

```python

import re
 
s = 'python.programming'
 
# without using \
match = re.search(r'.', s)
print(match)
 
# using \
match = re.search(r'\.', s)
print(match)
```

**.  Dot**

The dot (.) is a metacharacter in regular expressions that matches all single characters with the exception of the newline character (n).

For example, the regular expression a.b will match any string that has a "a" as its first character, followed by any single letter, and then a "b" as its last character. As a result, strings like "acb," "adb," "a1b," "a#b," etc. will be matched.

Similar to this, any string with at least two characters will match the regular expression.. The two dots signify that, with the exception of the newline character, any two characters may be present in the string. Hence, it will match characters such as "ab," "12," "!@," etc.


**Example**

Search for a sequence that starts with "Py", followed by three (any) characters, and an "n".

```python
import re
a = "Python programming"
#Search for a sequence that starts with "Py", followed by three (any) characters, and an "n":
x = re.findall("Py...n", a)
print(x)
```

**^  Caret**

The caret () is a metacharacter in regular expressions that identifies the start of a string. It is employed to determine whether a string contains a specific character or set of characters at the beginning.

The regular expression "c" will, for instance, match any string that begins with the letter "c," including "cat," "coal," "cow," and so on.

Similarly, the regular expression ^co will match any string that starts with the substring "co", such as "coal", "cow", etc.

Keep in mind that the caret () only serves to indicate the start of a string, not the start of each line in a multi-line string. With the multiline flag and the caret () metacharacter, one could match the start of each line in a multi-line string.


**Example**

Check if the string starts with 'hi'.

```python
import re
a = "hi how are you?"
#Check if the string starts with 'hi':
b = re.findall("^hi", a)
if b:
  print("Yes, the string starts with 'hi'")
else:
  print("No match")
```

**$  Dollar**

Dollar ($) is a metacharacter in regular expressions that identifies the end of a string. It is employed to determine whether a string contains a specific character or group of characters at the conclusion.

The regular expression s$, for instance, will match any string that ends with the letter "s," including "checks," "ends," "s," and so forth.

Similar to this, any string that ends with the substring "ks"such as "checks," "ks," etc.will be matched by the regular expression ks$.



**Example**

Check if the string ends with 'you'.

```python
import re
a = "Hi How are you"
#Check if the string ends with 'you':
b = re.findall("you$", a)
if b:
  print("Yes, the string ends with 'you'")
else:
  print("No match")
```

***  Star**

The star (*) metacharacter in regular expressions matches zero or more instances of the regex that comes before the * sign. When a specific character or group of characters may appear once, twice, or more in a pattern, it is used to match those patterns.

For instance, the regular expression xy*z will match any string that contains the letters "x" as its initial character, "y" zero or more times, and "z" as its last character. As a result, it will match strings such as "xz," "xy," "xyy," "dxy," etc. But since "y" is not followed by "z," the string "xydz" will not be matched.



**Example**

Search for a sequence that starts with "Py", followed by 0 or more  (any) characters, and an "m".

```python
import re
txt = "Python Programming"
#Search for a sequence that starts with "Py", followed by 0 or more  (any) characters, and an "m":
x = re.findall("Py.*m", txt)
print(x)
```

**+  Plus**

The plus (+) is a metacharacter used in regular expressions that matches one or more occurrences of the regex that comes before the + symbol. When a specific character or collection of characters must appear at least once in a pattern, it is used to match such patterns.

For instance, the regular expression xy+z will match any string that begins with the letter "x", continues with one or more instances of the letter "y", and ends with the letter "z". As a result, strings like "abc," "xyyz," "dxyz," etc. will be matched. But, as the string "xz" lacks a "y" character, it will not match. Similarly, it will not match the string "xydz" because "y" is not followed by "z".


**Example**

Search for a sequence that starts with "he", followed by 1 or more  (any) characters, and an "o"

```python
import re
a = "hello, what is this place?"
#Search for a sequence that starts with "he", followed by 1 or more  (any) characters, and an "o":
b = re.findall("he.+o",a)
print(b)
```

**?  Question Mark**

The question mark character in a regular expression really makes the previous character or group optional, meaning it can occur zero or one times. Thus in the example you supplied, the regex "ab?c" will match strings such as "ac", "abc", and "dabc" since the "b" is optional, but it will not match "abbc" because there are two "b"s and it will not match "abdc" because the "b" is not followed by "c".


**Example**

Search for a sequence that starts with "h", followed by 0 or 1  (any) character, and an "w".

```python
import re
a ="how are you"
#Search for a sequence that starts with "h", followed by 0 or 1  (any) character, and an "w":
x = re.findall("h.?w", a)
print(x)
```

**{}  Braces**

You can match the preceding character or group with a specified minimum and maximum number of repetitions by using the curly braces m,n in a regular expression.

For instance, the regex "ppq", "qpppc", and "pppqd" are examples of strings that the regex "p", "2, 3" will match. Nevertheless, it won't match strings that contain more than 3 "p" characters or less than 2 "p" characters (such as "p" or "qd").



**Example**

Search for a sequence that starts with "Wh", followed exactly 2 (any) characters, and an "e"

```python
import re
a = "Where are you heading?"
#Search for a sequence that starts with "Wh", followed excactly 2 (any) characters, and an "e":
b = re.findall("Wh.{2}e",a)
print(b)
```

**|  Or**

A regular expression's OR operator, represented by the pipe symbol |, matches either the pattern to the left of the pipe symbol or the pattern to the right.

The regex x|y, for instance, will match any string that contains either "x" or "y," such as "xrt," "ydf," "xytd," and so forth.


**Example**

Check if the string contains either "I" or "How".

```python
import re
a = "I love programming!"
#Check if the string contains either "I" or "How":
b = re.findall("I|how", a)
print(b)
if b:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**6. Special Sequences**

Special Sequences does not match for actual characters; instead, it displays the precise position in the supplied text. Each particular sequence has a distinct meaning and begins with the symbol.

Let's see what the special sequences and how to use them.



**\A**

No matter where it appears inside the pattern, A matches the beginning of the string. It matches a place in the string rather than a specific character.

For instance, the string "foo" would only match the regular expression pattern Afoo if it appeared at the very start of the input text. If "foo" appears later in the string, it would not match.




**Example**

Check if the string starts with "I".

```python
import re
a = "I love Python Programming"
#Check if the string starts with "I":
b = re.findall("\AI", a)
print(b)
if b:
  print("Yes, there is a match!")
else:
  print("No match")
```

**\b**

\b is used to match a word boundary, which is the position between a word character (as defined by \w) and a non-word character. a a d's s....

If the word begins with "string," b(string) matches the location right before the word rather than its beginning. Similarly, if the word ends in "string," (string)b fits the position right after the word. The parenthesis are optional and not required.

The word "cat" would only match, for instance, if it appeared as a whole word and not as a component of another word like "category" or "scatter," as in the case of the regular expression pattern \bcat\b.


**Example**

Check if "ing" is present at the beginning of a WORD.

```python
import re
a = "Python Programming"
#Check if "ing" is present at the beginning of a WORD:
b = re.findall(r"\bing", a)
print(b)
if b:
  print("Yes, there is at least one match!")
else:
  print("No match")

y= re.findall(r"ing\b", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")

```

**\B**

When a location is not a word boundary, the symbol B is used to match it. The location between two word characters or between two non-word characters is specifically what it matches.

This indicates that a pattern including the character B will only match if the specified characters are present inside of a word and not at its start or end.

For instance, if the word "cat" occurred inside a word like "scatter" or "category," the pattern \Bcat\B would match it, but not if it appeared at the start or end of the term.

**Example**

Check if "ing" is present, but NOT at the beginning of a word:

```python
import re
a = "Python Programming"
#Check if "ing" is present, but NOT at the beginning of a word:
y = re.findall(r"\Bing", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
#Check if "ain" is present, but NOT at the end of a word:
b = re.findall(r"ain\B",a)
print(b)
if b:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

After the d sequence, you can use the quantifier + to match one or more consecutive digits. In the input string, the pattern d+, for instance, might match one or more consecutive numbers.

**7. Sets**

A group of characters contained in square brackets is known as a set in regular expressions [].

Any character in a set matches another character in the set.


For instance, any one of the vowels "a," "e," I "o," or "u" matches the pattern [aeiou].

With a hyphen, we may additionally specify character ranges in a set, such as [a-z] to match any lowercase letter.

Moreover, you can negate a set by adding a caret right after the first bracket.

Any character that isn't in the set will match this. For instance, any character that is not a vowel matches the pattern [aeiou].


Let's see how to use various types of set patterns

**[pqr]**

Any character beginning with the letters "p," "q," or "r" falls into the character set or class "[pqr]".

The regular expression engine will try to match any one of the characters in the set at that location in the input string when it comes across a character set in a pattern.

By way of illustration, the pattern gr[ae]y would match either "grey" or "grey," but not "gruy" or "graay," because the character set [ae] matches either "a" or "e" at that place.


**Example**

Check if the string has any p, q, or r characters

```python
import re
txt = "Python Programming"
#Check if the string has any p, q, or r characters:
x = re.findall("[pqr]", txt)
print(x)
if x:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[p-z]**

Any lowercase letter between "p" and "z," inclusive, falls inside the character range "[p-z]".

The regular expression engine will try to match any one of the characters in the range at that place in the input string whenever it comes across a character range in a pattern.

For instance, because the character range [a-z] fits any lowercase letter at the second position in the string, the pattern b[a-z]g would match "bag," "bbg," "bkg," etc.

**Example**

Check if the string has any characters between p and z

```python
import re
a = "Python Programming"
#Check if the string has any characters between p and z:
y = re.findall("[p-z]", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[^pqr]**

Any character that is not "p," "q," or "r" matches the pattern [pqr], which is a negated character set.

The caret (caret) at the start of the character set negates it, making any character that isn't in the set match it.

For instance, the pattern [^aeiou] matches any non-vowel character, whereas the pattern [^0-9] matches any non-digital character.


**Example**

Check if the string has other characters than p, q, or r


```python
import re
a = "python programming"
#Check if the string has other characters than p, q, or r:
y = re.findall("[^pqr]", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[0123]**

A character set with the pattern [0123] can be used to represent any one of the numerals "0," "1," "2," or "3".

The regular expression engine will try to match any one of the characters in the set at that location in the input string when it comes across a character set in a pattern.

For instance, the pattern foo[0123]bar would match "foo0bar", "foo1bar", "foo2bar", or "foo3bar", but not "foo4bar" or "foo10bar".



**Example**

Check if the string has any 0, 1, 2, or 3 digits

```python
import re
a = "Python Programming 1"
#Check if the string has any 0, 1, 2, or 3 digits:
y = re.findall("[0123]",a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[0-9]**

The character range [0-9] corresponds to any one of the digits from "0" to "9".

The regular expression engine will try to match any one of the characters in the range at that place in the input string whenever it comes across a character range in a pattern.


For example, the pattern foo[0-9]bar would match "foo0bar", "foo1bar", "foo2bar", ..., "foo9bar", but not "foo10bar" or "fooa3bar".


**Example**

Check if the string has any digits

```python
import re
txt = "11:45 AM is the program"
#Check if the string has any digits:
x = re.findall("[0-9]", txt)
print(x)
if x:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[0-7][0-8]**

[0-7] Because the first character class [0-7] matches any digit from 0 to 7, and the second character class [0-8] matches any digit from 0 to 8, [0-8] would match any two-digit number between 00 and 78 inclusive. The pattern thus corresponds to integers from 00 through 78 inclusively but not to 79 or higher.

For example, the pattern [0-7][0-8] would match "00", "01", "02", ..., "77", and "78", but not "79" or "99".


**Example**

Check if the string has any two-digit numbers, from 00 to 78

```python
import re
a = "11:45 AM is the program"
#Check if the string has any two-digit numbers, from 00 to 78:
y = re.findall("[0-7][0-8]", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[a-zA-Z]**

A character range with the pattern [a-zA-Z] matches any letter from "a" to "z" in lowercase or uppercase.

Any character in the range at that location will be tried to match when the regular expression engine comes across this pattern in a string.

The pattern foo[a-zA-Z]bar, for instance, would match "fooabar", "foobbar",..., "foozbar", "fooAbar", "fooBbar",..., "fooZbar". The characters "foo1bar" and "foo#bar," however, do not fall inside the allowed range, therefore they would not be matched.

**Example**

Check if the string has any characters from a to z lower case, and A to Z upper case

```python
import re
a = "11:45 AM is the program"
#Check if the string has any characters from a to z lower case, and A to Z upper case:
y = re.findall("[a-zA-Z]", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**[+]**

A special character, such as +, *,., |, (), or $, loses its special meaning and is handled as a regular character when used within of a character set.

As a result, the pattern [+] would match any plus character "+" in the string but not any character sequences with a special meaning in regular expressions.

For instance, the pattern a+b would match any instance of the letters "a" and "b" together, matching "ab", "aab", "aaab", and so forth. The pattern [a+b] would, however, match every instance of the letter "a," the plus character "+," or the letter "b," matching "a," "+," "b," "ba," and so on.



**Example**

Check if the string has any + characters

```python
import re
a = "11:45 AM + is the program"
#Check if the string has any + characters:
y = re.findall("[+]", a)
print(y)
if y:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

**8.Match Object**

When you use the search() or match() functions in the Python re module to look for a pattern in a string, the output is a match object that includes details about the search and its result.

To learn more about the search result, you can use a number of the match object's methods and attributes.

You may acquire the position of a particular part of the matched string in the original string using these methods and attributes, and you can even save whole sections of the string for later use.

Let's see the implementation of various match objects

**.span()**

A match object's span() method returns a tuple containing the start and end positions of the match.


The start position is the index of the first character in the match, and the end position is the index of the character soon after the end of the match.


The match object's span() method, for instance, will return the tuple (6, 11) if you search for the pattern "world" using the phrase "hello world," indicating that the match begins at index 6 (the "w" character) and ends at index 11. (which is just after the "d" character).


**Example**

Match "programming" in the string pattern

```python
import re

string = "Python programming"
pattern = "programming"
match_object = re.search(pattern, string)

if match_object:
    print("Match found!")
    print("Match span:", match_object.span())
else:
    print("Match not found!")
```

**.string**

The string that was supplied to the function when the match object was formed is returned via the string attribute of a match object in the Python re module.


For instance, if you have a string called "hello world" and you search for the word "world," the string attribute of the match object will return the original string.

**Example**

Search "programming" in the string pattern


```python
import re

string = "Python programming"
pattern = "programming"
match_object = re.search(pattern, string)

if match_object:
    print("Match found!")
    print("Match string:", match_object.string)
else:
    print("Match not found!")
```

**.group()**

The section of the text where the match took place is returned by the group() method of a match object in Python's re module.


We can send an argument to the group() function to extract the section of the match that corresponds to a specific group if the pattern used to build the match object contains groups (i.e., portions of the pattern wrapped in parentheses).



**Example**

Search and match DOB in various match groups 

```python
import re

string = "James was born on 1891-02-10"
pattern = r"(\d{4})-(\d{2})-(\d{2})"
match_object = re.search(pattern, string)

if match_object:
    print("Match found!")
    print("Match group 0:", match_object.group(0))
    print("Match group 1:", match_object.group(1))
    print("Match group 2:", match_object.group(2))
    print("Match group 3:", match_object.group(3))
else:
    print("Match not found!")
```

**Summary**

* Regular Expressions (RegEx) is a powerful tool used for pattern matching and string manipulation. It is used to search and match specific patterns within a string of text. 

* In Python, RegEx can be implemented using the built-in re module. This module contains various functions that can be used to work with RegEx patterns, such as findall(), search(), split(), and sub().

* Metacharacters are special characters that have a special meaning in RegEx patterns. 

* Special sequences are shorthand codes that represent a specific character class or behavior in a RegEx pattern..

* Sets allow for matching a specific set of characters in a pattern.
